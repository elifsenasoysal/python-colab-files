```python
# prompt: b brandı için mart ayındaki yapılan satış sayılarını günlere göre çiz snscountplot ile

# b brandı için mart ayındaki yapılan satış sayılarını günlere göre çiz
plt.figure(figsize=(10, 6))
sns.countplot(x='Day_number', data=df[(df['Branch']=='B') & (df['Month']=='March')], palette='viridis')
plt.title('Mart Ayında B Şubesinin Günlere Göre Satış Sayıları')
plt.xlabel('Gün')
plt.ylabel('Satış Sayısı')
plt.show()
```
